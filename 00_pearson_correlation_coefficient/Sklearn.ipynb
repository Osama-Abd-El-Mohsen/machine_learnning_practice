```python
import pandas as pd
import math
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data_sets/Salary_Data.csv')
```

```python
data.head()
data.shape
print("="*50)
print(data.describe())
print("="*50)
print(data.info())
print("="*50)
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=10,train_size=0.8)

my_module = LinearRegression()
my_module.fit(x_train,y_train)
y_predict = my_module.predict(x_test)
my_module.score(x_train,y_train)
my_module.score(x_test,y_test)

plt.scatter(x_train,y_train,c='r')
plt.plot(x_train,my_module.predict(x_train))
plt.show()
```

```python
plt.scatter(x_test,y_test,c='r')
plt.plot(x_test,y_predict)
plt.show()
```

```python
plt.plot(y_test.values)
plt.plot(y_predict)
plt.show()
```

```python
plt.plot(abs(y_test.values - y_predict))
plt.show()
```

```python
mean_squared_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```

```python
print(f"coefficient  = {my_module.coef_}")
print(f"intercept  = {my_module.intercept_}")
```
